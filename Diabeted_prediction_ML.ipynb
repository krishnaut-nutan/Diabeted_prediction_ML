```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
db = pd.read_csv('diabetes.csv')
```

```python
# Separating Data and labels
X = db.drop(columns = 'Outcome', axis = 1)
Y = db['Outcome']
```

```python
X
```

```python
Y
```

```python
# Data standard standardize
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
st_data = scaler.transform(X)
```

```python
print(st_data)
```

```python
X = st_data
Y = db['Outcome']
```

```python
print(X, Y)
```

```python
# train test split
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Model
cl = svm.SVC(kernel='linear')
```

```python
cl.fit(X_train, Y_train)
```

```python
# Evaluation on training data
X_train_pred = cl.predict(X_train)
train_acc = accuracy_score(X_train_pred, Y_train)
```

```python
print(train_acc)
```

```python
# model evaluation for test data
X_test_pred = cl.predict(X_test)
test_acc = accuracy_score(X_test_pred, Y_test)
```

```python
print(test_acc)
```

```python
# Making a predictive system

input_data =(137,40,35,168,43.1,2.288,33,1)
# change input to numpy array
input_data_as_arr = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_arr.reshape(1, -1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)
```
